```python
import pandas as pd
import requests
import selenium
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
page=requests.get("https://www.imdb.com/india/top-rated-indian-movies/")
page.content
```

```python
soup=BeautifulSoup(page.content,"html.parser")
```

```python
Titel=[]
Rating=[]
Year_of_relese=[]
```

```python
Name=soup.find_all("td",class_="titleColumn")

Name[0:5]
```

```python
for i in Name:
    Titel.append(i.get_text().replace("\n",""))
Titel
```

```python
Year=soup.find_all("span",class_="secondaryInfo")
for i in Year:
    Year_of_relese.append(i.get_text().replace("\n",""))
```

```python
rating=soup.find_all("td",class_="ratingColumn imdbRating")
for i in rating:
    Rating.append(i.get_text().replace("\n",""))
Rating
```

```python
IMDB_INDIAN_100=pd.DataFrame({})
IMDB_INDIAN_100["Movie_Name"]=Titel[0:100]
IMDB_INDIAN_100["Year"]=Year_of_relese[0:100]
IMDB_INDIAN_100["Rarting"]=Rating[0:100]
IMDB_INDIAN_100
```

```python
name=[]
for i in range(len(IMDB_INDIAN_100["Movie_Name"])):
    temp=IMDB_INDIAN_100["Movie_Name"].iloc[i].split('(')
    try :
        
        name.append(temp[0].strip().upper())
    except :
        name.append('NONE')
IMDB_INDIAN_100["Movie_Name"]=name
```

```python
Name=[]
for i in range(len(IMDB_INDIAN_100["Movie_Name"])):
    temp=IMDB_INDIAN_100["Movie_Name"].iloc[i].split('.')
    try :
        
        Name.append(temp[1].strip().upper())
    except :
        Name.append('NONE')
IMDB_INDIAN_100["Movie_Name"]=Name
IMDB_INDIAN_100["Movie_Name"]
```
