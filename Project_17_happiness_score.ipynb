```python
#now convert the object to numeric with Lebel Encoder
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
cl=['Country','Region']
for i in cl:
    df[i]=le.fit_transform(df[i])
```

```python
#now detecting the outlier with zscore
from scipy.stats import zscore
z=np.abs(zscore(df))
print(np.where(z>3))
```

```python
#removing the Outlier
df_new=df[(z<3).all(axis=1)]
```

```python
#check the old data sets shape
df.shape
```

```python
#check the new data sets shape
df_new.shape
```

```python
#Lets divide dataset into input and output
df_x=df_new.drop(columns=["Happiness Score"])
y=df_new[["Happiness Score"]]
```

```python
print(df_x.shape)
print(y.shape)
```

```python
#skew ness of new data set
df_x.skew()
```

```python
#lets make a loop for skew data
for index in df_x.skew().index:
    if df_x.skew().loc[index]>0.55:
        df_x[index]=np.cbrt(df_x[index])
```

```python
df_x.skew()
```

```python
#lets scalling the data
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(df_x)
x=sc.transform(df_x)
x=pd.DataFrame(x,columns=df_x.columns)
print(x.shape)
print(y.shape)
```

```python
#make a function
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
def maxr2_score(regr,df_x,y):
    max_r_score=0
    for r_state in range(42,100):
        x_train, x_test, y_train, y_test = train_test_split(df_x, y,random_state = r_state,test_size=0.20)
        regr.fit(x_train,y_train)
        y_pred = regr.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        print("r2 score corresponding to ",r_state," is ",r2_scr)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r_state=r_state
    print("max r2 score corresponding to ",final_r_state," is ",max_r_score*100)
    return final_r_state
```

```python
#Build a function for cros_val_score
def kfold (model,x,y):
    c_scores=cross_val_score(model,x,y,cv=5,scoring="r2")
    cs=model,c_scores.mean()
    print("Mean r2 score for",cs)
    print("standard deviation in r2 score for",model,c_scores.std())
    print(c_scores)
```

```python
#using Decision Trees
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
maxr2_score(dtr,x,y)
```

```python
kfold(dtr,x,y)
```

```python
from sklearn.linear_model import LinearRegression
lnr=LinearRegression()
maxr2_score(lnr,x,y)
```

```python
kfold(lnr,x,y)
```

```python

#Lets chcek KNN
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV
knr=KNeighborsRegressor()
parameters = {"n_neighbors":range(2,30)}
rg = GridSearchCV(knr, parameters, cv=5,scoring="r2")
rg.fit(x, y)
rg.best_params_
```

```python
kn=KNeighborsRegressor(n_neighbors=4)
maxr2_score(kn,x,y)
```

```python
kfold(kn,x,y)
```

```python
from sklearn.svm import SVR
svr=SVR()
maxr2_score(svr,x,y)
```

```python
kfold(svr,x,y)
```

```python
#Based on above scores we will choose LinearRegression as our final algorithm
lnr=LinearRegression()
x_train, x_test, y_train, y_test = train_test_split(x, y,random_state = 68,test_size=0.20)
lnr.fit(x_train,y_train)
y_pred = lnr.predict(x_test)
```

```python
#Lets find the rmse and r2_score using sklearn.metrics for Linear Regressor model
print("MSE is: ",mean_squared_error(y_test,y_pred))
print("r2_score is: ",r2_score(y_test,y_pred)*100)
```
