# Data Preprocessing

```python
#converting objective to numeric
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
c=df_covid.columns
for i in c :
    if df_covid[i].dtypes=='object':
        df_covid[i]=l.fit_transform(df_covid[i])
       
```

# Outliers

```python
#Removing outliers
from scipy.stats import zscore
z=np.abs(zscore(df_covid))
df_covid_final=df_covid[(z<3).all(axis=1)]
print(df_covid.shape,df_covid_final.shape)
```

```python
d=df_covid_final.columns
for i in d:
    if df_covid_final[i].skew()>0.55:
        df_covid_final[i]=np.cbrt(df_covid_final[i])
        
df_covid_final['Hospitalization_Rate']=np.log(df_covid_final['Hospitalization_Rate'])  
df_covid_final.skew()
```

```python
#split the input and out put variable.
df_x=df_covid_final.drop(columns="Hospitalization_Rate",axis=1)
y=df_covid_final[["Hospitalization_Rate"]]
print(df_x.shape,y.shape)
```

# Data Standardize

```python
#scalling the data with standard scaler
from sklearn.preprocessing import MinMaxScaler
sc= MinMaxScaler()
x=sc.fit_transform(df_x)
x=pd.DataFrame(x,columns=df_x.columns)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=10)
poly.fit_transform(x)
x=pd.DataFrame(x,columns=df_x.columns)
```

# Model Creation

```python
#In this data set we know that out target variable Hospitalization_Rate is continuous so we are going to take Linear regresson
#lets make a function for getting the best random_satae for a model toget better accuracy(r2_score)
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
def max_acc(rgr,x,y):
    max_acc=0
    for r in range(42,100):
        x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r,test_size=0.20)
        rgr.fit(x_train,y_train)
        y_prd=rgr.predict(x_test)
        rc=r2_score(y_test,y_prd)
        if rc>max_acc:
            max_acc=rc
            final_r=r
    print("max R2_ score coressponding to ",final_r,"is",max_acc*100)
    print("MSE is: ",mean_squared_error(y_test,y_prd))
    
```

```python
#Build a function for cros_val_score
from sklearn.model_selection import cross_val_score
def kfold (model,x,y):
    c_scores=cross_val_score(model,x,y,cv=10,scoring="r2")
    cs=c_scores.mean()
    print("Mean r2 score for",cs)
    print("standard deviation in r2 score for",model,c_scores.std())
    print(c_scores)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
max_acc(lr,x,y)
```

```python
kfold(lr,x,y)
```
